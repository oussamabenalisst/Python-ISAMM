# Exercice 1 :
1. Génération du signal
Crée un signal sonore composé de deux sinusoïdes :
- une fréquence grave f1 = 440 Hz
- une fréquence aiguë f2 = 2000 Hz
- durée : 2 secondes
- fréquence d’échantillonnage Fs = 44100 Hz

```python
#Crée un signal sonore composé de deux sinusoïdes
import numpy as np
import soundfile as sf
f1 = 440  # Fréquence de la première sinusoïde
f2 = 2000  # Fréquence d'échantillonnage

duree=2
fs=44100
t=np.arange(0,duree,1/fs)
signal1=np.sin(2*np.pi*f1*t)
signal2=np.sin(2*np.pi*f2*t)
signal=signal1+signal2
sf.write('signal.wav', signal, fs)
```

* Explique le rôle de np.arange et la différence entre les deux composantes du signal.
    - np.arange crée un tableau de valeurs allant de 0 à duree avec un pas de 1/fs, représentant le temps.
    - La première composante du signal est une sinusoïde de fréquence f1 (440 Hz), tandis que la deuxième est une sinusoïde de fréquence f2 (2000 Hz).

```python
notes = [12, 15, 14, 10, 18, 20, 16, 14, 17, 19]
moy=np.mean(notes)
print(f"moyenne : {moy}")
print(f"note max : {np.max(notes)}")
print(f"note min : {np.min(notes)}")
print(f"ecart type : {np.std(notes)}")
print(f"variance : {np.var(notes)}")
#Ajoutez 2 points à toutes les notes supérieures à la moyenne with numpy
notes = np.array(notes)
notes[notes > moy] += 2
print("notes modifiées :", notes)
```
